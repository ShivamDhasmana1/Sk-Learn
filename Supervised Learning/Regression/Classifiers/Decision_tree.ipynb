# Decision Tree

### Decision Tree is a supervised machine learning algorithm used for classification and regression tasks. It works by repeatedly splitting the data based on feature values to form a tree-like structure of decisions and outcomes. Decision Trees are easy to understand and interpret, handle both numerical and categorical data, and require little data preprocessing. In machine learning, they are widely used in credit scoring, medical diagnosis, customer segmentation, and rule-based decision systems, though they can overfit unless properly pruned.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("Decision_Tree_data.csv"))

X = df[["Age","Income","Credit_Score","Employment_Status","Existing_Loan"]]
y = df["Loan_Approved"]
```

```python
le = LabelEncoder()

X_encoded = X.copy()
X_encoded['Employment_Status'] = le.fit_transform(X_encoded["Employment_Status"])
X_encoded['Existing_Loan'] = le.fit_transform(X_encoded['Existing_Loan'])

y_encoded = le.fit_transform(y)

model = DecisionTreeClassifier()
model.fit(X_encoded, y_encoded)

# Get user input
age = float(input("Enter your Age: "))
income = float(input("Enter your Income: "))
credit_score = float(input("Enter your credit Score: "))
employment_status = input("What is your Employment Status? (Employed, Unemployed or Self-Employed): ")
existing_loan = input("Is there any Existing Loan? (Yes or No): ")

# Encode categorical inputs
employment_encoded = le.fit_transform([employment_status])[0]
loan_encoded = le.fit_transform([existing_loan])[0]

new = pd.DataFrame({
    "Age": [age],
    "Income": [income],
    "Credit_Score": [credit_score],
    "Employment_Status": [employment_encoded],
    "Existing_Loan": [loan_encoded]
})

prediction = model.predict(new)[0]

if prediction:
    print("Lone Approved")
else:
    print("Lone not Approved")
```
