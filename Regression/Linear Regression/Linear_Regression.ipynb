# Linear Regression

### Linear Regression is a supervised machine learning algorithm used to model the relationship between a dependent variable and one or more independent variables by fitting a straight line. It works by minimizing the error between predicted and actual values, usually using the least squares method. Linear regression is widely used for prediction and trend analysis. In machine learning, it is commonly applied in areas like price prediction, sales forecasting, performance analysis, and risk estimation due to its simplicity, interpretability, and efficiency.

```python
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Study_hours.csv")
df = pd.DataFrame(data)
X = df[["Hours_Studied"]]
y = df["Marks_Obtained"]
```

```python
model = LinearRegression()
model.fit(X, y)
hours = float(input("Enter Hours Studied: "))
Prediction = model.predict([[hours]])
```

```python
if Prediction > 100:
    print(f"According to your {hours} hours Studying, your marks can be:\n 100")
else:
    print(f"According to your {hours} hours Studying, your marks can be:\n {np.round(Prediction,2)}")
```
