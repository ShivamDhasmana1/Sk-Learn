# K-Nearest Neighbour

### K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for classification and regression. It works by finding the K closest data points to a new input using a distance metric like Euclidean distance and making predictions based on majority voting (classification) or averaging (regression). KNN is simple, intuitive, and does not require model training, making it useful for pattern recognition, recommendation systems, and anomaly detection. However, it can be slow and memory-intensive for large datasets and is sensitive to feature scaling.


```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
df = pd.DataFrame(pd.read_csv("KNN_data.csv"))

X = df[["Age" , "Annual_Income" , "Spending_Score"]]
y = df["Purchased"]
```

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(X,y)

new = pd.DataFrame({
    "Age": [float(input("Enter Age "))],
    "Annual_Income": [float(input("Enter Annual Income "))],
    "Spending_Score": [float(input("Enter your Spending Score "))]
})

# [0] extracts the first element from the prediction array - model.predict() returns a numpy array, so [0] gets the actual prediction value
prediction = model.predict(new)[0]
```

```python
if prediction:
    print("Likely, to Purchase a BMW")
else:
    print("Likely, to not purchase a BMW")
```
