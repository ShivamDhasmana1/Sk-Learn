# Logistic Regression

### Logistic Regression is a supervised machine learning algorithm used for classification problems, especially binary classification. It models the probability of an outcome using the sigmoid (logistic) function, which maps values between 0 and 1. Instead of predicting continuous values, it predicts the class label based on a probability threshold. In machine learning, logistic regression is widely used in spam detection, medical diagnosis, fraud detection, and customer churn prediction because it is simple, efficient, and easy to interpret.


```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.DataFrame(pd.read_csv("Logistic_data.csv"))
X = df[["Study_Hours","Attendance","Previous_Score"]]
y = df["Passed"]
```

```python
model = LogisticRegression()
model.fit(X,y)

new_student = pd.DataFrame({
    "Study_Hours": [float(input("Enter Hours Studied "))],
    "Attendance": [float(input("Enter Attendance "))],
    "Previous_Score": [float(input("Enter your previous score "))]
})

# [0] extracts the first element from the prediction array - model.predict() returns a numpy array, so [0] gets the actual prediction value
Prediction = model.predict(new_student)[0]
print(Prediction)
```

```python
print(f"prediction is :")
if Prediction:
    print("Passed")
else:
    print("Failed")
```
