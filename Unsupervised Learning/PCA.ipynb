# Principal Component Analysis (PCA)

#### Principal Component Analysis (PCA) is an unsupervised dimensionality reduction technique used to reduce the number of features while preserving most of the data’s variance. It works by transforming correlated variables into a new set of uncorrelated components called principal components, ordered by the amount of variance they explain. PCA is widely used for data visualization, noise reduction, and improving model efficiency.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv("pca_practice_dataset.csv")
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df)
print(scaled)
```

```python
pca = PCA(n_components=2)
result = pca.fit_transform(scaled)
new_df = pd.DataFrame(result,columns=["PCA1", "PCA2"])

explained_varience = pca.explained_variance_ratio_
print("Varience captured by each PCA Component: ")
print(np.round(explained_varience*100,2))
```

### Plotting

```python
plt.Figure(figsize=(6,5))
plt.scatter(new_df["PCA1"], new_df["PCA2"])
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.grid(True)
plt.show()

print(new_df)
```
