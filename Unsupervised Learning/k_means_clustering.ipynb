# K-mean Clustering

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

### K-Means clustering is an unsupervised learning algorithm that groups data points into K clusters based on similarity.
### It works by minimizing the distance between data points and their respective cluster centroids.
### K-Means is commonly used for customer segmentation, pattern discovery, and data grouping.

```python
df = pd.read_csv("kmeans_customer_data.csv")
```

```python
X = df[["Age","Annual_Income","Spending_Score"]]

model = KMeans(n_clusters=3,random_state=42,n_init=10)
```

### random_state = same input → same output every time (for debugging, comparison, and fairness).

### n_init controls how many times K-Means is run with different starting points to ensure better and more stable clustering results.

### What does random_state = 42 mean?

* random_state is just a seed value used to control randomness.

* 42 itself has no special mathematical meaning — it’s simply a commonly used number.

* Using random_state=42 ensures you get the same result every time you run the cod
### Why specifically 42?

* It’s a popular convention in ML and programming (a geek reference from The Hitchhiker’s Guide to the Galaxy).

* You could use any number (0, 1, 100, 999) — the result would still be reproducible.

* What matters is consistency, not the value.

```python
df["Group"] = model.fit_predict(X)
```

```python
print(df)
```

## Plotting

```python
plt.figure(figsize=(6,5))
for group in df["Group"].unique():
    group_data = df[df["Group"] == group]
    plt.scatter(group_data["Age"], group_data["Annual_Income"], s =group_data["Spending_Score"]*2, label = f"Group {group}")
plt.xlabel("Age")
plt.ylabel("Annual_Income")
plt.legend()
plt.grid(True)
plt.show()

print(df)
```
